### Problem Statement

Before any bank gives a loan to a customer, it is important analysing at a number of assumptions. Therefore, it is not enough to evaluate the current scenario of the customer or the difficulties that the customer is going through. Some questions like for what purpose this customer will use the money or credit history they will help define how much money the bank will lend and what interest rates will apply.
Thus, after conducting all the analysis of the customer's finances, the bank will set out in an informed way how much to lend, according to what the customer will be able to pay monthly. That is, the bank must lend proportionally to the current conditions of the customer or the customer will risk falling back and becoming defaulter.
Loans are related to several factors like buying a property, car, personal or even repayment of another loan. A due to these factors the banks cross information that may be useful for the evaluation of the credit concession or not for the customer. Annual salary, loan terms, what is the reason for the loan, which are credit histories and among others. 

Purpose of the project is to make analyzes in Databricks with scripts in Python and Spark to answer the questions if there is relation between the currencies features and the groups created along the data engineering. 

Expected outcomes are to have results through statistical analyzes and machine learning that demonstrate if there is a correlation between the currencies variables and the groups. Checking whether the variables of these groups will influence the results.

####  PCA - Principal Componente Analysis

This is a method whose basic purpose is to analyze the data used for its reduction, elimination of overlaps and the choice of the most representative forms of data from linear combinations of the original variables. PCA is a way of identifying the relationship between characteristics extracted from the data.

##### 
I did the PCA of the variables 'RevLineCr_Num2', 'LowDoc_Num2', 'UrbanRural' and 'NewExist' in relation to the currencies variales 'DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv' and 'SBA_Appv', following:

```
# PCA of Variable 'RevLineCr_Num2'
# Import PCA library 
from sklearn.decomposition import PCA
# spliting input and target

df_PCA_LowDoc_Num2 = df.copy() # copy from dataset original to new variable

X = df_PCA_LowDoc_Num2.drop(['LoanNr_ChkDgt', 'Name', 'City', 'State', 'Zip', 'Bank', 'BankState',
       'NAICS', 'ApprovalDate', 'ApprovalFY', 'Term', 'NoEmp', 'NewExist',
       'CreateJob', 'RetainedJob', 'FranchiseCode', 'UrbanRural', 'RevLineCr',
       'LowDoc', 'ChgOffDate', 'DisbursementDate',  'MIS_Status', 
       'RevLineCr_Num', 'LowDoc_Num',  'LowDoc_Num2'],1)  # Drop the columns that are not necessary for the PCA 

Y = df.RevLineCr_Num2 

pca = PCA(n_components=2).fit_transform(X) # pca output is an array
pca_df = pd.DataFrame(pca,columns=['pca_1','pca_2']) # transforming back to df
pca_Y = pd.concat([pca_df, Y],axis=1) # concat the columns 'pca_df' and 'Y'

import matplotlib.pyplot as plt
import seaborn as sns
sns.FacetGrid(pca_Y, hue="RevLineCr_Num2", palette="Set2", size=6).map(plt.scatter, "pca_1", "pca_2").add_legend()
display()
```
